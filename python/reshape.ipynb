```python
import numpy as np
```

```python
num = 2 * 3 * 4 * 5
a1 = np.array(range(num))
print(a1)
```

```python
a2 = a1.reshape(2,3,4,5)
print(a2)
```

np.reshapeは、指定する次元の1つを-1にできる。-1を指定された次元は、残りの次元から自動で推測される

```python
a3 = a1.reshape(2,3,4,-1)
print(a3)
```

```python
a4 = a1.reshape(2,3,-1,5)
print(a4)
```

# reshape時の順番を検討

```python
print(a2.reshape(5,4,3,-1))
```

### Q: reshapeしてから転置するのと、最初から転置した次元を指定したら同じ結果？

```python
print(a2.reshape(2*3*4, -1).T)
```

```python
print(a2.reshape(-1, 2*3*4))
```

A: 違う
