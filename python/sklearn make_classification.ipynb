```python
from sklearn.datasets import make_classification

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
X, y = make_classification()
```

```python
print(X.shape, y.shape)
```

```python
cors = [np.corrcoef(X[:,i], y)[0,1] for i in range(X.shape[1])]
plt.plot(cors)
```
