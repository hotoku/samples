```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
fig = plt.figure(figsize=(5,6))
ax = fig.add_subplot(111)
ax.boxplot(range(10), labels=["a"])
plt.show()
```

```python
fig = plt.figure(figsize=(5,6))
a1 = fig.add_subplot(121)
a2 = fig.add_subplot(122)
a1.boxplot(range(10), labels="a")
a2.boxplot(range(20), labels="b")
plt.show()
```

```python
df = pd.DataFrame({"x":range(10), 
                   "y":range(10,20)})
print(df)
```

```python
fig = plt.figure(figsize=(5,6))
ax = fig.add_subplot(111)
ax.boxplot(df)
plt.show()
```

boxplotにdfを直接入れると、↑こんな感じに・・・

```python
fig = plt.figure(figsize=(5,6))
ax = fig.add_subplot(111)
ax.boxplot([df["x"], df["y"]], labels=df.columns)
plt.show()
```
