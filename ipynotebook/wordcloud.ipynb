```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
freq = {
    "ほげ": 1,
    "ふが": 2    
}
wc = WordCloud(
    font_path="~/Library/Fonts/ipaexg.ttf", # ipaexgフォントは別途インストールする。方法はググる
    width=300, 
    height=200,
    background_color="white",
    min_font_size=10
).generate_from_frequencies(freq)
plt.imshow(wc)
```
