GeoJSONのポリゴンにPopupを載せる

```python
import folium
from h3 import h3
import geojson as gj
```

```python
center = h3.geo_to_h3(35.5, 139.5, 11)
codes = h3.k_ring(center, 10)

def swap(p):
    return [p[1], p[0]]

def code2polygon(code):
    vtx = h3.h3_to_geo_boundary(code)
    poly = folium.GeoJson(dict(
        type = "Polygon",
        coordinates = [[swap(p) for p in vtx]]
    ))
    poly.add_child(folium.Popup(code))
    return poly
    
polys = list(map(code2polygon, codes))
```

```python
fmap = folium.Map(location=(35.5, 139.5), zoom_start=20)
for p in polys:
    p.add_to(fmap)
folium.LatLngPopup().add_to(fmap)
fmap
```
