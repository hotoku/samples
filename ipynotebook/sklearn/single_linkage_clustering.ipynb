```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_classification
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
np.random.seed(0)
ps, lab = make_classification(n_features=2, n_redundant=0, n_clusters_per_class=1, class_sep=2)
```

```python
df = pd.DataFrame(ps)
df["label"]=lab
```

```python
est = AgglomerativeClustering(n_clusters=None, linkage="single", distance_threshold=1)
sns.scatterplot(x=df[0], y=df[1], hue=est.fit_predict(ps))
```
