```python
def make_color(mix):
    red = random.random()
    green = random.random()
    blue = random.random()

    if (mix is not None):
        red = (red + mix.rgb_r) / 2;
        green = (green + mix.rgb_g) / 2;
        blue = (blue + mix.rgb_b) / 2;

    return sRGBColor(red, green, blue)
```

```python
mix = sRGBColor(1,1,1)
cols = [make_color(mix) for i in range(20)]
```

```python
from h3 import h3
import folium
```

```python
cols[0].get_rgb_hex()
```

```python
center = h3.geo_to_h3(35, 139, 11)
k_ring = h3.k_ring(center, 10)
fmap = folium.Map(
    location = (35, 139),
    zoom_start = 20
)

def swap(p):
    return p[1], p[0]

for i, cell in enumerate(k_ring):
    p = folium.GeoJson(
        dict(
            type="Polygon",
            coordinates=[[
                swap(p) for p in h3.h3_to_geo_boundary(cell)
            ]],
            id=i
        ),
        style_function=lambda f: dict(
            color = cols[f["geometry"]["id"] % len(cols)].get_rgb_hex(),
            weight = 0,
            fillOpacity = 0.8
            # color = "#ffffff"
        )
    )
    p.add_to(fmap)
fmap
```
