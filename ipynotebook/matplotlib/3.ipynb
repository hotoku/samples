```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.lines as lines
```

```python
fig, ax = plt.subplots()
l2 = lines.Line2D([0,1,2], [0,1,0])
ax.set_xlim(0, 3)
ax.set_ylim(0, 3)
ax.add_line(l2)
```

`Line2D`で折れ線を引ける

```python
fig, ax = plt.subplots()
l2 = lines.Line2D([0,1,2], [0,1,0], marker=".", markersize=20)
ax.set_xlim(0, 3)
ax.set_ylim(0, 3)
ax.add_line(l2)
```

マーカーを置ける。ただし、サイズ・種類は1種類しか指定できない

```python
fig, ax = plt.subplots()
l2 = lines.Line2D([0,1,2], [0,1,0])
ax.set_xlim(0, 3)
ax.set_ylim(0, 3)
ax.plot(1,1, "bo")
ax.plot(2,1, "ob")
ax.plot(0,1, "ro", markersize=10)
ax.plot(1,2, marker=".", markersize=20)
ax.plot(1,2, marker=".", markersize=20, color="y")
ax.plot(2,2, marker="x", markersize=20, color="y")
ax.add_line(l2)
```

マーカーを個別に置きたかったら`ax.plot`を使う
