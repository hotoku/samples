```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y = np.arange(100, 110)
x,y
```

```python
z = np.tensordot(x, y, axes=0)
z
```

```python
plt.contour(x, y, z)
```

```python
plt.contourf(x, y, z)
```

# levels引数を試す

数値を渡すと階調をコントロールできる

```python
plt.contourf(x, y, z, levels=100)
```

```python
plt.contourf(x, y, z, levels=3)
```

刻み幅を直接に渡すこともできる

```python
plt.contourf(x, y, z, levels=np.arange(1201, step=300)) # 1200だと、右上が白くなる
```

```python
plt.contourf(x, y, z, levels=np.arange(1001, step=50))
```

```python
plt.contourf(x, y, z, levels=[0, 3, 10, 30, 100, 300, 1000])
```

levelsの範囲を、値の範囲の外まで広げると、色の範囲もコントロールできる

```python
plt.contourf(x, y, z, levels=np.arange(2001, step=50))
```
