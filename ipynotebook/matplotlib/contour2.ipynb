```python
import matplotlib.pyplot as plt
import numpy as np
import itertools as it
from matplotlib import cm
from matplotlib.cm import ScalarMappable
from matplotlib.colors import Normalize

%matplotlib inline
```

```python
def make_xyz(n, s):
    x, y = np.linspace(-10, 10, n), np.linspace(-10, 10, n)
    z = np.array([
        np.exp(-(a**2+b**2)/s**2) for a, b in it.product(x, y)
    ]).reshape(n, n)
    return x, y, z
```

```python
cmap = cm.get_cmap("jet")
norm = Normalize(0, 1)
mble = ScalarMappable(cmap=cmap, norm=norm)
```

```python
sigma=1
```

## 解像度が荒い場合

```python
x, y, z = make_xyz(10, sigma)
```

### levelsを与えない

```python
fig, ax = plt.subplots()
con = ax.contourf(x, y, z, cmap=cmap)
plt.colorbar(mble)
```

### levelsを与える

```python
fig, ax = plt.subplots()
con = ax.contourf(x, y, z, levels=np.linspace(0,1,100), cmap=cmap)
plt.colorbar(mble)
```

## 解像度が高い場合

```python
x, y, z = make_xyz(100, sigma)
```

### levelsを与えない

```python
fig, ax = plt.subplots()
con = ax.contourf(x, y, z, cmap=cmap)
plt.colorbar(mble)
```

### levelsを与える

```python
fig, ax = plt.subplots()
con = ax.contourf(x, y, z, levels=np.linspace(0,1,100), cmap=cmap)
plt.colorbar(mble)
```
