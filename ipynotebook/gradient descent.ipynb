```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def f(x, y):
    a = 0.2
    return a * x**2 + y**2, (a*2*x, 2*y)
```

等高線を描く

```python
x = np.linspace(-10, 10, 40)
y = np.linspace(-3, 3, 40)
X, Y = np.meshgrid(x, y)
Z, _ = f(X, Y)
plt.contour(X, Y, Z)
```

勾配降下法の実装

```python
def gd(f, ini, lr, num):
    x, y = ini
    vals = np.zeros(num)
    args = np.zeros((num+1, 2))
    args[0,:] = x,y
    for i in range(num):
        v, d = f(x, y)
        x -= lr * d[0]
        y -= lr * d[1]
        vals[i] = v
        args[i+1,:] = x,y
    return vals, args
```

勾配降下法を実行

```python
lr, num = 0.1, 1000
vals, args = gd(f, (8, 1), lr, num)
fig, ax = plt.subplots(figsize=(10,3))
ax.contour(X, Y, Z)
ax.plot(args[:,0], args[:,1])
ax.scatter(args[:,0], args[:,1])
```

```python
lr, num = 1, 1000
vals, args = gd(f, (8, 1), lr, num)
fig, ax = plt.subplots(figsize=(10,3))
ax.contour(X, Y, Z)
ax.plot(args[:,0], args[:,1])
ax.scatter(args[:,0], args[:,1])
```
